```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic=sns.load_dataset('titanic')
```

```python
titanic.head(1)
```

Count plot

```python
titanic['class'].value_counts()
```

```python
plt.figure(figsize=(7,5))
sns.countplot(x='class',data=titanic)
```

```python
plt.figure(figsize=(7,5))
sns.countplot(x='class',data=titanic, palette='Set2') # isme hum color add krte he usinig palette
```

```python
plt.figure(figsize=(7,5))
sns.countplot(x='class',data=titanic, hue='who', palette='hls') # iss me hum categories dal sakete he (hue)
```

```python
plt.figure(figsize=(7,5))
sns.countplot(x='class',data=titanic, hue='who', palette='hls',edgecolor='black')
```

## Bar plot

```python
plt.figure(figsize=(7,5))
sns.barplot(x='class',y='survived',data=titanic)  # categorical features nikalne ke liye use krte he
```

```python
plt.figure(figsize=(7,5))
sns.barplot(x='class',y='survived',data=titanic, hue="who", palette="Set2")
```

```python
## vaha ke black bar ka mem line ka mtlb he ki vaha kuchh outliers he / ya outlires liye gaye  he
```

Box plot

```python
### it is use for the for the outliers detection
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(x='class',y='fare',data=titanic, hue="who", palette="viridis")
```

Violin plot

```python
plt.figure(figsize=(7,5))
sns.violinplot(x='class',y='fare',data=titanic,palette="viridis")
```

Historgram

```python
# for contineous feature distribution
```

```python
plt.figure(figsize=(7,5))
sns.histplot(titanic['fare'],palette="Set2",bins=50,alpha=0.5,kde='true')  #### kde="ture" plot me line ko drow krta he for show the smooth overall distribution of the table
```

kde =kernel density estimator

```python
## kde = kernel density estimator
```

```python
sns.kdeplot(titanic['fare'], fill='true')
```

Heatmap-correlation matrix

```python
correlation_matrix=titanic.corr(numeric_only=True)
```

```python
correlation_matrix
```

```python
plt.figure(figsize=(7,5))
sns.heatmap(correlation_matrix,annot=True) ## annot=True enhances the informativeness and clarity of the heatmap
```

Scatter plot -  numerical vs numerical

```python
sns.scatterplot(x='age',y='fare',data=titanic)
```

Regression plot

```python
sns.regplot(x='age',y='fare',data=titanic) ## it show the regression whether or classification
```

Subplot

```python
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
sns.scatterplot(x='age',y='fare',data=titanic, hue="who",palette="Set2")

plt.subplot(1,2,2)
sns.kdeplot(data=titanic,x="fare",palette="Set2",)
```
