Bar Plot

```python
gender_counts=titanic['sex'].value_counts()
```

```python
gender_counts
```

```python
plt.figure(figsize=(6,4))
plt.bar(gender_counts.index,gender_counts.values,
        color=["blue","red"])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution')
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.bar(gender_counts.index,gender_counts.values,
        color=["blue","red"], alpha=0.5) ##for color theme
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution')
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.bar(gender_counts.index,gender_counts.values,
        color=["blue","red"], alpha=0.5, edgecolor="indigo") # for the border color
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution')
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.bar(gender_counts.index,gender_counts.values,
        color=["blue","red"], alpha=0.5, edgecolor="indigo") # for the border color
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution')
plt.show()
plt.tight_layout() # for proper arrangment of the data like,(Avoid Overlaping)
```

```python
plt.savefig('gender_distribution.png') # for saving the file
```

```python
plt.figure(figsize=(6,4))
plt.barh(gender_counts.index,gender_counts.values,
        color=["blue","red"], alpha=0.5, edgecolor="indigo") # .barh for horizontal bar
plt.ylabel('Gender') # here we change the axis
plt.xlabel('Count')
plt.title('Gender Distribution')
plt.show()
plt.tight_layout() # for proper arrangment of the data like,(Avoid Overlaping)
```

Histogram

```python
titanic['age'].isnull().sum()
```

```python
age=titanic['age'].dropna()
```

```python
age.isnull().sum()
```

```python
  plt.figure(figsize=(6,4))
  plt.hist(age, bins=30, color='blue', edgecolor='yellow') # in bins me hum aaur number of data print krte he
  plt.xlabel('Age')
  plt.ylabel('Frequency')
  plt.title('Age Distribution')
  plt.grid(True) ## isme hum grid ko draw krte he
  plt.show()
```

Pie chart

```python
survived_counts=titanic['survived'].value_counts()
```

```python
survived_counts  # o- died , 1-alive
```

```python
plt.figure(figsize=(6,6))
plt.pie(survived_counts.values,labels=["Died","Survived"])
```

```python
plt.figure(figsize=(6,6))
plt.pie(survived_counts.values,labels=["Died","Survived"], autopct='%1.1f%%') ## percent me represent krne me help krta he
plt.title('Survived vs Died')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(survived_counts.values,labels=["Died","Survived"], autopct='%1.1f%%', startangle=90) ## startangle me hum data ka angle defined krte he , jo kis angle dikhna chahiye
plt.title('Survived vs Died')
plt.legend()
plt.show()
```

Scatter Plot - Bivariate- Num VS Num

```python
titanic[["age","fare"]].isnull().sum()
```

```python
data=titanic.dropna(subset=["age","fare"]) ## jitene age ke kiye utne hi fare ke kiye
```

```python
data
```

```python
data[["age","fare"]].isnull().sum()
```

```python
data[["age","fare"]].shape
```

```python
data[["age","fare"]].notnull().sum()
```

```python
data.head()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data["age"],data["fare"])
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data["age"],data["fare"],color="Indigo" ,alpha=0.5, marker="o") #marker se hum shape change kr sakte he
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare')
```

Box plot - Outlier Dtetection

```python
plt.figure(figsize=(6,4))
plt.boxplot(data["age"],)
plt.ylabel('Age')
plt.title('Box Plot of Age')
plt.show() # above 60 yo 80 outliers he
```
