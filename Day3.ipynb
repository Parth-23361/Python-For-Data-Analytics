Loops

```python
fruits=['apple','banana','cherry']
```

```python
for fruit in fruits:
  print(fruit)
```

```python
# in keyword always use for iteration
# iteration mean one by one accessing the elements
```

```python
for i in range(1,11):
  print(i)
```

```python
total = 0
for i in range(1,11):
  total = total + i
  #total += i
print(total)
```

```python
num=5
for i in range(1,11):
  print(f"{num} x {i} = {num*i}")
```

```python
for i in range(1,11):
  print("*"*i)
```

```python
#nested loop for 1 to 3 table
for i in range(1,4):
  for j in range(1,6):
    print(f"{i} x {j} = {i*j}")
    print()
```

```python
# While loop
```

```python
i=1
while i<=10:
  print(i)
  i+=1
  #i=i+1
```

```python
#reverse the number
```

```python
num=123
rev=0
while num>0:
  rev=rev*10+ num% 10
  num=num//10
print(rev)
```

```python
usename= 'admin'
password='admin123'
user= str(input('enter your username'))
passw= str(input('enter your password'))
if usename == user and password == passw:
  print('login successfully')
else:
  print('invalid username or password')
```
